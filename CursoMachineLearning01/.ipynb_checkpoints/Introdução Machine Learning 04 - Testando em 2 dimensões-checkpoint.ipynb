```python
import numpy as np
import pandas as pd 

#Projetos finalizados: 1
#Projetos não finalizados: 0

#Lê os dados csv
uri = "https://gist.githubusercontent.com/guilhermesilveira/1b7d5475863c15f484ac495bd70975cf/raw/16aff7a0aee67e7c100a2a48b676a2d2d142f646/projects.csv"
dados = pd.read_csv(uri)

#Faz o mapeamento para traduzir o cabeçalho das colunas
mapa = {
    "unfinished":"nao_finalizados",
    "expected_hours":"horas_esperadas",
    "price":"preco"
}

#Faz o mapeamento para que Finalizados = 1 e Não finalizados = 0 (para deixar mais intuitivo)
troca = {
    0 : 1,
    1 : 0
}

#Formata o cabeçalho dos dados de acordo com o mapeamento
dados = dados.rename(columns = mapa)
dados['finalizados'] = dados.nao_finalizados.map(troca)

dados.head()
```

```python
import seaborn as sns

#Exibe graficamente a distribuição de projetos
sns.scatterplot(x="horas_esperadas", y="preco", hue="finalizados", data=dados)
```

```python
#Exibe graficamente a distribuição de projetos
sns.relplot(x="horas_esperadas", y="preco", col="finalizados", hue="finalizados", data=dados)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score 


SEED = 20 

#Elimina a necessidade de usar random_state = SEED no train_test_split e no LinearSVC
np.random.seed(SEED)
x = dados[['horas_esperadas', 'preco']]
y = dados['finalizados']

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25, stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

#modelo = LinearSVC()
modelo = SVC()

modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)

#DÚVIDA: POR QUE A ACURÁCIA ESTÁ VARIANDO SE ESTOU USANDO RANDOM_STATE = SEED?
#Nesse caso, é preciso usar o Random_state = seed também no modelo LinearSVC e não apenas no train_test_split.
```

```python
#Análise dos dados de teste: 
sns.relplot(x="horas_esperadas", y="preco", hue=teste_y, data=teste_x)
```

```python
#Note que o x (horas esperadas) vai de 0 a 100 e o y (preco) vai de 0 a 25000 no gráfico.
x_min = teste_x.horas_esperadas.min()
x_max = teste_x.horas_esperadas.max()

y_min = teste_y.preco.min()
y_max = teste_y.preco.max()

print(x_min, x_max, y_min, y_max)
```
