```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(".\\introducao-a-data-science-aula0\\aula0\\dados.csv")
data.head()
```

```python
data[['Idade', 'Anos de Estudo', 'Renda', 'Altura']].describe().round(2)
```

```python
#0 - Masculino
#1 - Feminino
percentual = pd.DataFrame({'Percentual (%)': data.Sexo.value_counts(normalize=True).round(2) * 100})
percentual = percentual.rename(index = {0:'Masculino', 1:'Feminino'})
percentual
```

```python
#8 - Indigena
#2 - Branca
#4 - Negra
#0 - Parda 
#2 - Nao declarado

data.Cor.value_counts()
percentual_etnia = pd.DataFrame({'Percentual (%)': data.Cor.value_counts(normalize=True).round(2) * 100})
percentual_etnia.rename(index = {8:'Negra', 2:'Parda', 4:'Amarela', 0:'Branca', 6:'Nao Declarado'}, inplace=True)
percentual_etnia
```

```python
sexo = {0: 'Masculino',
        1: 'Feminino'}

cor = {0: 'Indígena',
        2: 'Branca',
        4: 'Preta',
        6: 'Amarela',
        8: 'Parda',
        9: 'Sem declaração'}


frequencia_etnia = pd.crosstab(data.Sexo, data.Cor, aggfunc='mean', values=data.Renda)
frequencia_etnia
```

```python
frequencia_etnia.rename(index = sexo, inplace=True)
frequencia_etnia.rename(columns = cor, inplace=True)
frequencia_etnia.round(2)
```
